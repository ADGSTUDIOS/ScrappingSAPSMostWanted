```python
from bs4 import BeautifulSoup
import urllib.request
requests = urllib.request.urlopen('https://www.saps.gov.za/crimestop/wanted/list.php').read()
soup = BeautifulSoup(requests)
```

## Scrapping Criminal links from SAPS Database

```python
links = []
for anchor in soup.findAll('a'):
        link = 'https://www.saps.gov.za/crimestop/wanted/'+anchor['href']
        links.append(link)
        nameofsuspects = anchor.text

url = []   
for index,link in enumerate(links):
    if link[len(link)-1].isdigit():
        url.append(link)
```

Number of Records to Scrape



```python
totalrecordstoscrape = len(url)
totalrecordstoscrape
```

Viewing our urls

```python
url
```

## Now we scrape all the urls from the SAPS Database and data mine

```python
FullName = []
comment = []
Gender = []
CrimeDate = []
CaseNumber = []
count = 0
for websitelink in url:
  try:
    requests = urllib.request.urlopen(websitelink).read()
    soup = BeautifulSoup(requests)
    tds = soup.find_all('td')  
    Gender.append(tds[12].text)
    CrimeDate.append(tds[8].text)
    CaseNumber.append(tds[26].text)
    for anchor in soup.findAll('h2'):
      FullName.append(anchor.text)
    for anchor in soup.find("tr").findAll('p'):
      comment.append(anchor.text.strip())
    count+=1 
    print('New Record Added {0}, We now have {1} records captured'.format(count,FullName))
  except Exception as e:
    print('error',e)
```

## Database Report

```python
import pandas as pd

print("Scrapping success rate achieved : ",float(len(FullName)/totalrecordstoscrape)*float(100),'%')
print('We have {0} out of {1} records that we have scrapped'.format(len(FullName),totalrecordstoscrape))
```

### Our Database

```python
data = list(zip(FullName,comment,Gender,CrimeDate,CaseNumber))
df = pd.DataFrame(data, columns = ['CriminalName', 'Comment', 'Gender','CrimeDate','CaseNumber'])
df
```

#### Function to Insert Into SAPS Database using our stored procedure in MSSQL

```python
from time import sleep
from tqdm.auto import tqdm
```

```python
def InsertToSAPSDatabase(CriminalName,Comment,CrimeDate,Gender,CaseNumber):
  if Gender == 'Male' or Gender == 'Female':   
    db.ExecuteQuery("exec sp_InsertSAPSRecords @CriminalName = '{0}',@Comment = '{1}',@Gender = '{2}',@CrimeDate  = '{3}',@CaseNumber = '{4}'".format(CriminalName.replace("'",""),Comment.replace("'",""),Gender.replace("'",""),CrimeDate.replace("'",""),CaseNumber.replace("'","")))
  else:
    db.ExecuteQuery("exec sp_InsertSAPSRecords @CriminalName = '{0}',@Comment = '{1}',@Gender = '{2}',@CrimeDate  = '{3}',@CaseNumber = '{4}'".format(CriminalName.replace("'",""),Comment.replace("'",""),'Unknown',CrimeDate.replace("'",""),CaseNumber.replace("'","")))
```

```python
totalnumberofrecords = len(df)
try:
  print('Job Started ... \n')
  for x in tqdm(range(totalnumberofrecords)):
      InsertToSAPSDatabase(FullName[x],comment[x],CrimeDate[x],Gender[x],CaseNumber[x])

  print('Job Finished :-)')
except Exception as e:
  print('error failed to insert record:',e)
```
